# UNIVARIATE ANALYSIS:

```python
sns.set_style("whitegrid");
sns.FacetGrid(df,hue='Survival status',size=5).map(sns.distplot,"Age").add_legend();
plt.show();
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df,hue='Survival status',size=5).map(sns.distplot,"Year operation").add_legend();
plt.show();
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df,hue='Survival status',size=8).map(sns.distplot,"Axillary nodes detected").add_legend();
plt.show();
```

BY OBSERVING ABOVE GRAPHS,IT SEEMS  THAT AMONG THREE FEATURES,'AXILLARY NODES DETECTED' FEATURE IS CLASSIFYING BETTER THAN OTHERS.



BUT EVENTHOUGH,THERE IS A LOT OF OVERLAPPING PRESENT,WE HAVE TO GO FOR OTHER METHODS FOR ANALYSIS. 

```python
df_s1=df.loc[df["Survival status"]==1];
df_s2=df.loc[df["Survival status"]==2];
```

```python
counts,bin_edges=np.histogram(df_s1["Age"],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


counts,bin_edges=np.histogram(df_s2["Age"],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

IT IS CLEAR THAT CDFs OF 'AGE' ARE ALMOST COMPLETELY OVERLAPPING 

```python
counts,bin_edges=np.histogram(df_s1["Axillary nodes detected"],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


counts,bin_edges=np.histogram(df_s2["Axillary nodes detected"],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

```

BASED ON 'AXILLARRY NODES' ALSO, IT IS UNABLE TO DISTINGUISH

```python
counts,bin_edges=np.histogram(df_s1["Year operation"],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)



counts,bin_edges=np.histogram(df_s2["Year operation"],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```
